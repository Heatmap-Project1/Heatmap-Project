```python
LC = ['location.name', 'location.region', 'location.country', 'location.lat',
       'location.lon', 'current.last_updated', 'current.temp_c', 'current.temp_f',
       'current.is_day', 'current.wind_mph', 'current.wind_kph',
       'current.wind_degree', 'current.wind_dir', 'current.pressure_mb',
       'current.pressure_in', 'current.precip_mm', 'current.precip_in',
       'current.humidity', 'current.cloud', 'current.feelslike_c',
       'current.feelslike_f', 'current.windchill_c', 'current.windchill_f',
       'current.heatindex_c', 'current.heatindex_f', 'current.dewpoint_c',
       'current.dewpoint_f', 'current.vis_km', 'current.vis_miles',
       'current.uv', 'current.gust_mph', 'current.gust_kph',
       'current.will_it_rain', 'current.chance_of_rain',
       'current.will_it_snow', 'current.chance_of_snow',
       'current.air_quality.co', 'current.air_quality.no2',
       'current.air_quality.o3', 'current.air_quality.so2',
       'current.air_quality.pm2_5', 'current.air_quality.pm10',
       'current.air_quality.us-epa-index',
       'current.air_quality.gb-defra-index']

import time
import requests
import pandas as pd

ldf = []
url = 'http://api.weatherapi.com/v1/current.json'

# 1. Use a Session object to reuse the same TCP connection (massively reduces connection errors)
with requests.Session() as session:
    session.headers.update(headers)

for i in dn:
        for j in dn[i]:
            if pd.isna(j):
                continue
                
            city = str(j)
            state = str(i)
            
            params = {
                'key': 'e55105ea9e214004b65173640261207',
                'q': f"{city},{state}",
                'aqi': 'yes'
            }
            
            try:
                # Use the session instead of raw requests.get
                li = session.get(url, params=params, timeout=10)
                
                if li.status_code == 200:
                    di = pd.json_normalize(li.json())[LC]
                    ldf.append(di)
                elif li.status_code == 429:
                    print("⚠️ Hitting rate limits! Pausing for 5 seconds...")
                    time.sleep(5)
                else:
                    print(f"Skipping {city}, {state}: HTTP Error {li.status_code}")
                    
            except requests.exceptions.RequestException as e:
                print(f"Connection dropped for {city}, {state}. Pausing briefly... Error: {e}")
                time.sleep(2) # Give the network a moment to recover if it drops
                
            # 2. Add a tiny breath pause (0.2 seconds) to respect the API limits
            time.sleep(0.2)

df = pd.concat(ldf, ignore_index = True)#.to_csv('weather.csv')
```

```python
#conversion formula
'''
west = -180
east = 180
north = 90
south = -90
^ for equirectangular map

px = ((lon-west)/(east-west))*imagewidth
py = ((north-lat)/(north-south))*imageheight

round pixel values
'''
```

```python
import pandas as pd
```

```python
dnow = pd.read_csv('finalweather34k.csv')
```

```python
lat, long = dnow['location.lat'], dnow['location.lon']
```

```python
dpixel = pd.concat([lat,long], axis = 1)
```

```python

west = -180
east = 180
north = 90
south = -90
w = iblank.size[0]
h = iblank.size[1]
#^ for equirectangular map
'''
px = ((lon-west)/(east-west))*imagewidth
py = ((north-lat)/(north-south))*imageheight
'''
#round pixel values
```

```python
dpixel['px'], dpixel['py'] = np.round(((long-west)/(east-west))*w).astype(int), np.round(((north-lat)/(north-south))*h).astype(int)
```

```python
dpixel
tlist = dnow['current.temp_c']
M,m = tlist.max(),tlist.min()
denom = M-m
dpixel['TCOLOR'] = tlist.apply(lambda x: np.round(((x-m)/denom)*255).astype(int))
#np.concat([dpixel, tlist, tcolor], axis = 0)
#dpixel = pd.concat([dpixel, tlist], axis = 1)
dpixel['px'] = dpixel['px'].clip(0, w-1)
dpixel['py'] = dpixel['py'].clip(0, h-1)
```

```python
np.max(dpixel['px']), np.max(dpixel['py']),w ,h
```

```python
for i in np.array(dpixel[['px','py', 'TCOLOR']]):
    iblank.putpixel((int(i[0]),int(i[1])), (int(i[2]),0,0))
```
